```python

tarefas = []


def listar_tarefas():
    print("Lista de Tarefas:")
    for i, tarefa in enumerate(tarefas, 1):
        status = "[x]" if tarefa["concluida"] else "[ ]"
        print(f"{i}.{status} {tarefa['descricao']}")


def registrar_tarefa(descricao):
    nova_tarefa = {
        "descricao": descricao.capitalize(),
        "concluida": False
    }
    tarefas.append(nova_tarefa)
    print("Tarefa registrada!!!")


def marcar_como_realizada(identificador):
    if identificador > 0 and identificador <= len(tarefas):
        tarefa = tarefas.pop(identificador - 1)
        tarefa["concluida"] = True
        tarefas.insert(0, tarefa)
        print("Tarefa marcada como realizada!")
    else:
        print("Identificador inválido ou tarefa já realizada.")


def editar_tarefa(identificador, nova_descricao):
    if identificador > 0 and identificador <= len(tarefas):
        tarefas[identificador - 1]["descricao"] = nova_descricao.capitalize()
        print("Tarefa editada!")
    else:
        print("Identificador inválido.")


while True:
    print("\n--- MENU ---")
    print("1. Listar tarefas")
    print("2. Adicionar nova tarefa")
    print("3. Marcar tarefa como realizada")
    print("4. Editar tarefa")
    print("5. Sair")

    opcao = input("Escolha uma opção: ")

    if opcao == "1":
        listar_tarefas()
    elif opcao == "2":
        descricao = input("Digite a descrição da nova tarefa: ")
        registrar_tarefa(descricao)
    elif opcao == "3":
        identificador = int(input("Digite o identificador da tarefa a ser marcada como realizada: "))
        marcar_como_realizada(identificador)
    elif opcao == "4":
        identificador = int(input("Digite o identificador da tarefa a ser editada: "))
        nova_descricao = input("Digite a nova descrição da tarefa: ")
        editar_tarefa(identificador, nova_descricao)
    elif opcao == "5":
        break
    else:
        print("Opção inválida. Tente novamente.")
```
